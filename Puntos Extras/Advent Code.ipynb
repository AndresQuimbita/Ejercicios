```python
def boards(A, B):
    for n in A:
        for b in B:
            for i in b:
                for idx, val in enumerate(i):
                    if val == n:
                        i[idx] = 0
                        break
                suma = 0
                if win(b):
                    for i in b:
                        suma += sum(i)    
                    return suma*n
def win(board):
    win = True
    for i in board:
        for j in range(len(i)):
            if i[j] != 0:
                win = False
                break
            else:
                win = True
        if win:
            return True
    
```

```python
def last_boards(A, B):
    boardWinners = [False for x in range(len(B))]
    sums = []
    last = []
    for n in A:
        for b in range(len(B)):
            if boardWinners[b]:
                continue
            for i in B[b]:
                for idx, val in enumerate(i):
                    if val == n:
                        i[idx] = 0
                        break                    
            if win(B[b]):
                suma = 0
                for i in B[b]:
                    suma += sum(i)    
                boardWinners[b] = True
                sums.append(suma)
                last.append(n)
    sums.pop()
    last.pop()
    return sums[-1]*last[-1]
def win(board):
    win = True
    for i in board:
        for j in range(len(i)):
            if i[j] != 0:
                win = False
                break
            else:
                win = True
        if win:
            return True
    
```

```python
class Cave:
    def __init__(self, Name = "", Big = None, Other = [] ):
        self.Name = Name
        self.Big = Big
        self.Other = Other
```

```python
class Path:
    def __init__(self, path = [], visited = {}, visitedSmall = False ):
        self.path = path
        self.visited =  visited
        self.visitedSmall = visitedSmall
        
```

```python
def copy_map(a):
    out = {}
    for items in a.items():
        out[items[0]] = items[1]
    return out
        
def paths_1(A):
    queue = [Path(["start"])]
    paths = []
    while queue: 
        cur = queue.pop(0) 
        cave = A[cur.path[-1]]
        if cave.Name == 'end':
            paths.append(cur.path)
            continue
        
        visited = copy_map(cur.visited)
        if not cave.Big:
            visited[cave.Name] = True
        for i in cave.Other:
            if i in cur.visited:
                if cur.visited[i]:
                    continue
            path = cur.path.copy()
            path.append(i)                
            queue.append(Path(path, visited))
    return paths    
    
    
```

```python
class Step:
    def __init__(self, pairs = {}, count = {}):
        self.pairs = pairs
        self.count = count
    
```

```python
class Output:
    def __init__(self, pair1 = "", pair2 = "{}", element = ""):
        self.pair1 = pair1
        self.pair2 = pair2
        self.element = element
```

```python
def step_1(s, dic):
    out = Step({},{})
    for i in s.count:
        out.count[str(i)] = s.count[i]
    for i in s.pairs.items():
        t = dic[str(i[0])]
        if str(t.pair1) in out.pairs:
            out.pairs[str(t.pair1)] += i[1]
        else:
            out.pairs[str(t.pair1)] = i[1]
        if str(t.pair2) in out.pairs:
            out.pairs[str(t.pair2)] += i[1]
        else:
            out.pairs[str(t.pair2)] = i[1]
        if str(t.element) in out.count:
            out.count[str(t.element)] += i[1]
        else:
            out.count[t.element] = i[1]
    
    return out
```

```python
def newStep(inp):
    s = Step({},{})
    for i in range(len(inp)-1):
        pair = inp[i:i+2]
        if str(pair) in s.pairs:
            s.pairs[str(pair)] += 1
        else:
            s.pairs[str(pair)] = 1
        
    for i in range(len(inp)):
        if str(inp[i]) in s.count:
            s.count[str(inp[i])] += 1
        else:
            s.count[str(inp[i])] = 1
    return s
```
