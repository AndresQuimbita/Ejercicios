### 3.  La búsqueda en un árbol binario de búsqueda es siempre mas rápida que la búsqueda lineal en un arreglo

Falso, a pesar de que la complejidad de un Binary Search Tree es O(log(n)), ya que el árbol está ordenado, hay casos en que el Binary Search Tree no es más rápida que la búsqueda lineal, como por ejemplo, cuando no esta balanceada, sino que solo tiene un solo camino hacia la derecha o la izquierda, que cumpla con la definición de BST, y el recorrido en ese caso, el peor de los casos es O(n), lo cual es la misma complejidad que en la búsqueda lineal de un arreglo, por ello, no siempre BST es más rápido que la búsqueda lineal en un arreglo.

### 5.  Un algoritmo de complejidad Ω(nlog(n)) es mas rápido que un algoritmo de complejidad Ω(n)


Falso, debido a que nlog(n) crece más que n, entonces hay k*(nlong(n)) $\leq$ f(n) y como se ve en la gráfica, nlog(n) se aleja de de constante 1 que deseamos, en cambio n se esta más abajo, entonces k*(n) $\leq$ f(n). nlog(n) $\geq$ n, entonces k(n)$\leq$k(nlog(n))$\leq$ f(n). Por lo tanto, la complejidad $\Omega$(n) es más rápida que $\Omega$(nlog(n)). El mejor caso de f(n) es nlog(n), la cual es más lento que n.

````python
 def reverseList(self, head):
    if head == None or head.next == None:
        return head
    prev, curr = None, head
    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
    return prev
````

### Implemente los métodos enqueue y decreaseKey, tal que ambos métodos tengan complejidad O(log(n)), sin utilizar funciones nativas de Python. Por ejemplo, si necesita hacer una búsqueda, no use la función de Python que implementa esto, programe la función usando for o while loops.

````python
def binary_insertion_sort(alist):
        for i in range(1, alist.item_count):
            temp = alist.queue[i]
            pos = binary_search(alist, temp, 0, i) + 1

            for k in range(i, pos, -1):
                alist.queue[k] = alist.queue[k - 1]

            alist.queue[pos] = temp

def binary_search(alist, key, start, end):
        if end - start <= 1:
            if key[0] > alist.queue[start][0]:
                return start - 1
            else:
                return start

        mid = (start + end)//2
        if alist.queue[mid][0] > key[0]:
            return binary_search(alist, key, mid, end)
        elif alist.queue[mid][0] < key[0]:
            return binary_search(alist, key, start, mid)
        else:
            return mid
````

````python
import ctypes
class PriorityQueue(object):
    """
    Implementation of the queue data structure
    """

    def __init__(self, n):
        self.item_count = 0
        self.n = n
        self.queue = self._create_queue(self.n)        
    
    def _create_queue(self, n):
        """
        Creates a new stack of capacity n
        """
        return (n * ctypes.py_object)()

    def dequeue(self):

        c = self.queue[0]

        for i in range(1,self.item_count):

            self.queue[i-1] = self.queue[i]

            self.queue[self.item_count - 1] = ctypes.py_object

            self.item_count -= 1

        return c
    def enqueue(self, item):
            if self.item_count == self.n:
                raise ValueError("no more capacity")
            self.queue[self.item_count] = item
            self.item_count += 1
            binary_insertion_sort(self)
    def decreaseKey(self,index,value_decrease):
        if value_decrease > self.queue[index][0]:
            raise ValueError("Value decrease invalid. Is bigger than key value.")
        self.queue[index]= (self.queue[index][0]-value_decrease,self.queue[index][1])
        binary_insertion_sort(self)
````
