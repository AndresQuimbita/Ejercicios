##  1. Sort the following functions in decreasing order of asymptotic complexity ($O(f(n))$):

- $f_1(n) = n^\sqrt{n}$
- $f_2(n) = 2^n$
- $f_3(n) = {n\choose 2}$
- $f_4(n) = \sum_{i=2}^n (i-1)$

```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import scipy.special
%matplotlib inline  
```

```python
f = pd.DataFrame()
f['n'] = np.arange(0,20,0.5)
f['n^(n)^1/2'] = np.power(f['n']*np.ones(f.shape[0]), f['n']**1/2)
f['2^n'] = np.power(2*np.ones(f.shape[0]), f['n'])
f['(n 2)'] = np.power(scipy.special.binom(f['n'], 2),1)
f['sum'] = np.power(((f['n']**2)-f['n'])//2,1)
```

```python
f.plot(x='n', y=['n^(n)^1/2','2^n','(n 2)','sum'])#, logy=True)
plt.ylim([0,60])
plt.xlim([0,15])
plt.xlabel('n')
plt.grid()
```

Como se puede ver en la gráfica, la funcion $n^\sqrt{n}$ sobre pasa a la funcion $2^n$ en un punto determinado, y esta a su vez sobrepasa a las otras dos funciones ${n\choose 2}$ y $\sum_{i=2}^n (i-1)$, y aunque no se vea en el gráfico, las dos últimas funciones $f_3$ y $f_4$ son equivalentes y son $O(f(n))$ entre si, por lo que se le puede poner una encima de la otra o viceversa. El orden quedaria:
- $n^\sqrt{n}>2^n>{n\choose 2}\geq\sum_{i=2}^n (i-1)$

```python
def insertion_sort_descreasing(array):
    """
    Implementation of insertion sort
    """
    s = 1000000*time.time()
    for i in range(1, len(array)):        
        key = array[i]
        j = i-1
        while (j >= 0) and (key > array[j]):
            array[j+1] = array[j]
            j -= 1
        array[j+1] = key
    return 1000000*time.time()-s
```

## 5. CLRS 3.1-2
pg. 52

Se tiene que $(n+a)^b$ = $\Theta (n^b)$ y se quiere demostrar que es verdad $\forall a,b $, donde $b>0$ y $a,b$ son constantes.

Se plantea que $(n+a)\leq 2n$, cuando $|a|\leq n$ y $n+a \geq \frac{n}{2}$, cuando $|a|\leq \frac{n}{2}$. Entonces:

$\frac{n}{2} \leq (n+a) \leq 2n$

Como queremos llegar a la expresion $n^b$, entonces se elva todo a b

$(\frac{n}{2})^b \leq (n+a)^b \leq 2^b n^b$

$(\frac{1}{2})^b n^b \leq (n+a)^b \leq 2^b n^b$

De modo que se c1=$(\frac{1}{2})^b$, c2=$2^b$, hay constantes que cumplan con la regla de Big-Theta, por lo tanto $(n+a)^b$ = $\Theta (n^b)$ es verdadero

## 6. CLRS 3.1-6
pg. 53

Si recurrimos a la definicion de $\Theta (g(n))$ que se plantea: $\Theta(g(n)) = \{f(n) : \exists c_1, c_2, n_0 > 0, \text{such that } 0 \leq c_1g(n) \leq f(n) \leq c_2g(n), \forall n \geq n_0 \}$.

Esto muestra que para una funcion $g(n)$ existen constantes, $c_1$ y $c_2$, de modo que se cumplan las desigualdades. Sabiendo que  $O(g(n))$ es peor caso y $\Omega(g(n))$ es el mejor caso, sabemos su definción de modo que: 
- $O(g(n)) = \{f(n) : \exists c, n_0 > 0 \text{ such that } 0 \leq f(n) \leq cg(n), \forall n \geq n_0 \}$
- $\Omega(g(n)) = \{f(n) : \exists c, n_0 > 0 \text{ such that } 0 \leq cg(n) \leq f(n), \forall n \geq n_0 \}$

Con estas definiciones podemos ver, para el peor caso se tiene la desigualdad  $0 \leq f(n) \leq cg(n)$, la cual si nos fijamos en la definicion de $\Theta$ es igual a la primera parte de la misma y lo mismo sucede con el mejor caso, $0 \leq cg(n) \leq f(n)$, si juntamos ambas inecuaciones obtenemos, $0 \leq cg(n) \leq f(n) \leq f(n) \leq cg(n) =>0 \leq c1g(n) \leq  f(n) \leq c2g(n)$, y esto nos lleva a la definicion de Big-Theta.

## 7. CLRS 3-4 (a,b,e,g)

a) $f(n) = O(g(n))$ implies $g(n)=O(f(n))$

Es no es cierto debido a que para las funciones $f(n)=\frac{n}{2}$ y $g(n)=6^n$ no se cumple lo establecido.Por definicion de Big-O sabemos que $0 \leq f(n) \leq cg(n)$, por lo que si resolvemos:

- $0 \leq f(n) \leq cg(n)$
- $\frac{n}{2} = \frac{1}{2} n$, de modo que $\frac{1}{2}$ es la consntante c de la inecuacion expresada arriba
- Entonces, $0 \leq f(n) \leq \frac{1}{2}g(n)$ es verdadero

- $0 \leq g(n) \leq cf(n)$
- $6^n = 2^n*3^n$, en este caso no hay una constante que cumpla la inecuacion
- Entonces, $0 \leq f(n) \leq g(n)$ es falso

Concluyendo que $f(n) = O(g(n))$ implies $g(n)=O(f(n))$ es falso.


b)$f(n)+g(n) = \Theta(min(f(n),g(n)))$

Por definicion de Big-O sabemos que $0 \leq f(n) \leq cg(n)$, entonces $0 \leq f(n)+g(n) \leq c*min(f(n),g(n))$, si ponemos un ejemplo con $f(n)=2^n$ y $g(n)=8^n$, en donde se sabe que $2^n$ crece mas lento que $5^n$ resolvemos que: 
- $0 \leq f(n)+g(n) \leq c*min(f(n),g(n))$
- $2^n+8^n = 2^n+2^n*4^n = 2^n(1+4^n)$, en el cual no se puede encontrar una constante que cumpla la desigualdad.
- Entonces, $0 \leq f(n)+g(n) \leq c*min(f(n),g(n))$ es falso



e) $f(n) = \Theta((f(n))^2)$

Por definicion de Big-Theta sabemos que $\Theta(g(n)) = \{f(n) : \exists c_1, c_2, n_0 > 0, \text{such that } 0 \leq c_1g(n) \leq f(n) \leq c_2g(n), \forall n \geq n_0 \}$. Si consideramos a $f(n)= 1/n^4)$, entonces tenemos que:
- $0 \leq c_1f(n)^2 \leq f(n) \leq c_2f(n)^2$
- $0 \leq c_1\frac{1}{n^{16}} \leq \frac{1}{n^4} \leq c_2\frac{1}{n^{16}}$
- $\frac{1}{n^4} \leq c_2\frac{1}{n^{16}}$, si suponemos que n = kc2, donde k es constante, entonces:
- $\frac{1}{(kc_2)^4} \leq c_2\frac{1}{(kc_2)^{16}}=\frac{1}{k^{16}c_2^{15}}$, lo cual deja una contradiccion, ya que mientras mayor es denominador, mas pequeño es el resultado, por lo tanto  $\frac{1}{(kc_2)^4} \leq c_2\frac{1}{(kc_2)^{16}}$ es falso y el enunciado tambien.


g) $f(n) = \Theta(f(n/2))$

Si tomamos con ejemplo $f(n)= 2^n$, y analizamos parte de la inecuacion que hace que Big-Theta sea posible, tenemos:
- $0 \leq c_1f(\frac{n}{2}) \leq f(n) \leq c_2f(\frac{n}{2})$
